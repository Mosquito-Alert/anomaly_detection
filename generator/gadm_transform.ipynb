# GADM Transform

Trasform the original world GADM geopackage file: remove columns, filter by country, etc.

```python
import pandas as pd
import geopandas as gpd
import os
```

```python
# * Base directory.
DATA_DIR = os.path.join(os.getcwd(), 'data')
```

# Load and Filter

```python
INPUT_DIR = os.path.join(DATA_DIR, 'input', 'geo', 'GADM', 'gadm_410.gpkg')
gadm4_world_gdf = gpd.read_file(INPUT_DIR)
gadm4_esp_gdf = gadm4_world_gdf[gadm4_world_gdf['GID_0'] == 'ESP']
gadm4_esp_gdf.head(5)
```

```python
gadm4_esp_gdf.columns
```

# GDF1. To generate bites data

```python
# ? Use 'GID_4' OR 'UID/fid' as the unique identifier????
gadm4_esp_simplified_gdf = gadm4_esp_gdf[['UID', 'NAME_4', 'geometry']].rename(
    columns={'NAME_4': 'name'}
).reset_index(drop=True)
gadm4_esp_simplified_gdf
```

```python
# * Save the filtered GeoDataFrame to a new file.
OUTPUT_DIR = os.path.join(DATA_DIR, 'output', 'GADM')
os.makedirs(OUTPUT_DIR, exist_ok=True)

gadm4_esp_simplified_gdf.to_file(
    os.path.join(OUTPUT_DIR, 'gadm_410_esp_simplified', 'gadm_410_esp_simplified.shp'),
    driver='ESRI Shapefile',
    index=False,
    encoding='utf-8'
)
gadm4_esp_simplified_gdf.to_file(
    os.path.join(OUTPUT_DIR, 'gadm_410_esp_simplified.gpkg'),
    driver='GPKG',
    index=False,
    encoding='utf-8'
)
```

# GDF2. To save it in DB

```python
columns_to_keep = [
        'UID', 'CONTINENT', 'GID_0', 'NAME_0', 'VARNAME_0',
        'GID_1', 'NAME_1', 'VARNAME_1', 'NL_NAME_1', 'TYPE_1', 'ENGTYPE_1',
        'GID_2', 'NAME_2', 'VARNAME_2', 'NL_NAME_2', 'TYPE_2', 'ENGTYPE_2',
        'GID_3', 'NAME_3', 'VARNAME_3', 'NL_NAME_3', 'TYPE_3', 'ENGTYPE_3',
        'GID_4', 'NAME_4', 'VARNAME_4', 'TYPE_4', 'ENGTYPE_4',
        'GID_5', 'NAME_5', 'TYPE_5', 'ENGTYPE_5',
        'geometry'
    ]

# Original columns:
# See https://gadm.org/metadata.html
# 'UID', 'GID_0', 'NAME_0', 'VARNAME_0',
# 'GID_1', 'NAME_1', 'VARNAME_1', 'NL_NAME_1', 'ISO_1', 'HASC_1', 'CC_1', 'TYPE_1', 'ENGTYPE_1', 'VALIDFR_1',
# 'GID_2', 'NAME_2', 'VARNAME_2', 'NL_NAME_2', 'HASC_2', 'CC_2', 'TYPE_2', 'ENGTYPE_2', 'VALIDFR_2',
# 'GID_3', 'NAME_3', 'VARNAME_3', 'NL_NAME_3', 'HASC_3', 'CC_3', 'TYPE_3', 'ENGTYPE_3', 'VALIDFR_3',
# 'GID_4', 'NAME_4', 'VARNAME_4', 'CC_4', 'TYPE_4', 'ENGTYPE_4', 'VALIDFR_4',
# 'GID_5', 'NAME_5', 'CC_5', 'TYPE_5', 'ENGTYPE_5',
# 'GOVERNEDBY', 'SOVEREIGN', 'DISPUTEDBY', 'REGION', 'VARREGION', 'COUNTRY', 'CONTINENT', 'SUBCONT', 'geometry'
```

```python
gadm4_esp_clean_gdf = gadm4_esp_gdf[columns_to_keep].reset_index(drop=True)
gadm4_esp_clean_gdf.columns = gadm4_esp_clean_gdf.columns.str.lower()
gadm4_esp_clean_gdf
```

```python
# * Save the filtered GeoDataFrame to a new file.
OUTPUT_DIR = os.path.join(DATA_DIR, 'output', 'GADM')
os.makedirs(OUTPUT_DIR, exist_ok=True)

gadm4_esp_clean_gdf.to_file(
    os.path.join(OUTPUT_DIR, 'gadm_410_esp_clean.gpkg'),
    driver='GPKG',
    index=False,
    encoding='utf-8'
)
```
